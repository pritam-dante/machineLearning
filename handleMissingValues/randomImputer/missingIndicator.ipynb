# its better to use simpleImputer class with add_indicator attribute

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer,MissingIndicator
```

```python
df = pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.sample(5)
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape , X_test.shape
```

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf,y_train)

y_pred = clf.predict(X_test_trf)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### add 1% in the accuracy

## easy and better practice is using simple imputer only
# here it goes

```python
si_final = SimpleImputer(add_indicator=True)
```

```python
X_train_final = si_final.fit_transform(X_train)
X_test_final = si_final.transform(X_test)
```

```python
clf_final = LogisticRegression()
clf_final.fit(X_train_final,y_train)
y_pred_final = clf_final.predict(X_test_final)
accuracy_score(y_test,y_pred_final)
```
