# Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

### 1. Extract Year

```python
date['date_year'] = date['date'].dt.year
date.sample(5)
```

### 2. Extract Month

```python
date['date_month_no'] = date['date'] .dt.month
date.sample(3)
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.sample(3)
```

### Extract Days

```python
date['date_day'] = date['date'].dt.day
date.sample(3)
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek
date.sample(3)
```

```python
# day of week name
date['date_dow_name'] = date['date'].dt.day_name()
date.sample(3)
```

```python
date.drop(columns=['product_id','city_id','orders'])
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Saturday','Sunday']),1,0)
```

```python
date.sample(10)
```

### Etract week of the year

```python
date['date_week']= date['date'].dt.week

date.sample(5)
```

```python
# extract quater

date['quarter']= date['date'].dt.quarter
date.sample(5)
```

## Extract semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
```

```python
date.sample(10)
```

# Extract time elasped between dates

```python
import datetime

today = datetime.datetime.today()
```

```python
today
```

```python
today - date['date']
```

```python
(today-date['date']).dt.days
```

```python
# Months passed

np.round((today - date['date']) / np.timedelta64(1,'M'),0)
```
