Seaborn --> Data Visualization Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#inbuilt library
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for the plots --> without the theme there will be no grid
sns.set_theme()
```

```python
#visualize the data
sns.relplot(x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size',data = tips)
```

```python
#load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

ScatterPlot

```python
#house price dataset
from sklearn.datasets import load_boston
house_boston = load_boston()

house = pd.DataFrame(house_boston.data,columns = house_boston.feature_names)
house['price'] = house_boston.target
```

```python
house.head()
```

Distribution Plot

```python
sns.distplot(house['price'])
```

Correlation:
 1. Positive Correlation
 2. Negative Correlation

Heat Map

```python
correlation = house.corr()
```

```python
#construction a heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues' )
```
