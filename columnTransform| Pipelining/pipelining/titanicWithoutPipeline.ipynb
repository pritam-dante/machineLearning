```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.sample(4)
```

```python
# step! --> TrainTestSplit
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=2)
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
# Applying imputation
si_age = SimpleImputer() #default strategy = mean
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
X_train_age
```

```python
df.sample()
```

```python
# one hot Encoding Sex and Embarked
ohe_sex = OneHotEncoder(sparse=False,handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)
```

```python
X_train_embarked
#3 because 3 embarked categories
```

```python
X_train_sex
#for male and female
```

```python
X_train.head(2)
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_embarked,X_train_sex),axis =1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_embarked,X_test_sex),axis=1)
```

```python
X_train_transformed.shape
```

```python
X_train_transformed
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
